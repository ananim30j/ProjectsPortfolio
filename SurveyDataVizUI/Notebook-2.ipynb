# Analysis of Organizational Focus and Economic Impact on Creative Workers and Administrators

## Introduction
In this notebook, we aim to delve deep into the roles of Creative Workers and Arts Administrators, with a specific focus on understanding the primary arts focus of their associated organizations and its distribution. Additionally, we will explore the representation of funders within the dataset. The analysis will also provide insights into the economic challenges faced by these organizations, specifically in terms of their operating budgets and the extent of layoffs relative to the number of staff members.

We first read the dataset and print few rows for visualization

```python
import pandas as pd
# Load the dataset
df = pd.read_excel("2023-organized-survey.xlsx")

# Display the first few rows of the dataset
df.head()
```

We wanted to see what kind of art organizations associated with "Creative Workers" and "Arts Administrators" focus on. To do this:

We identified the exact descriptions for "Creative Workers" and "Arts Administrators" from our data.Using these descriptions, we picked out the relevant entries from our dataset.We then checked the main art focus of these organizations by looking at the "Primary Creative Focus of Your Organization" columnBy counting the different types of art focuses, we got a breakdown of the main art types these organizations are involved in.
This breakdown helps us understand what kind of art is popular among these organizations.

```python
# Detailed descriptions for the roles
roles_descriptions = [
    "Creative Worker (In-house graphic designers, copywriters, art directors employed at a business working in the creative industries)",
    "Arts Administrator (Theater managers, development staff, executive directors, program coordinators at an arts-based business usually a non-profit, e.g., at a museum, symphony, theater, etc.)"
]

# Filter the dataset based on the roles
filtered_df = df[df["Describe Your Primary Role in Creative Sector"].isin(roles_descriptions)]

# Analyze the "Primary Creative Focus of Your Organization" column to determine the primary arts focus and its distribution
arts_focus_distribution = filtered_df["Primary Creative Focus of Your Organization"].value_counts()

arts_focus_distribution
```

### Creative Workers, Arts Administrators: What is the primary arts focus of the organizations and at what distribution? What about funders?

```python
import matplotlib.pyplot as plt

# Define role descriptions
creative_worker_description = "Creative Worker (In-house graphic designers, copywriters, art directors employed at a business working in the creative industries)"
arts_admin_description = "Arts Administrator (Theater managers, development staff, executive directors, program coordinators at an arts-based business usually a non-profit, e.g., at a museum, symphony, theater, etc.)"

# Filter the dataset based on the roles
creative_worker_df = df[df["Describe Your Primary Role in Creative Sector"] == creative_worker_description]
arts_admin_df = df[df["Describe Your Primary Role in Creative Sector"] == arts_admin_description]

# Analyze the "Primary Creative Focus of Your Organization" column for each role
creative_worker_focus_distribution = creative_worker_df["Primary Creative Focus of Your Organization"].value_counts()
arts_admin_focus_distribution = arts_admin_df["Primary Creative Focus of Your Organization"].value_counts()

# Plot the distributions
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 16))

# Creative Worker plot
creative_worker_focus_distribution.plot(kind='barh', color='lightgreen', ax=axes[0])
axes[0].set_title('Distribution of Primary Creative Focus for "Creative Worker" Role')
axes[0].set_xlabel('Number of Organizations')
axes[0].set_ylabel('Primary Creative Focus')
axes[0].invert_yaxis()

# Arts Administrator plot
arts_admin_focus_distribution.plot(kind='barh', color='lightcoral', ax=axes[1])
axes[1].set_title('Distribution of Primary Creative Focus for "Arts Administrator" Role')
axes[1].set_xlabel('Number of Organizations')
axes[1].set_ylabel('Primary Creative Focus')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()


```

### Insights:

### Diverse Focus: 
The organizations associated with "Creative Worker" and "Arts Administrator" roles exhibit a diverse range of primary creative focuses.

### All-encompassing Approach: 
The category "All Arts, Comprehensive" has the highest representation, indicating that there are organizations that do not limit themselves to a single arts discipline but rather have a holistic approach encompassing various art forms.

### Design & Visual Arts: 
Both "Commercial Design" (encompassing architecture, industrial, and interior design) and "Visual Art" have significant representation. This suggests a strong inclination towards visual and design-centric disciplines in the sample.

### Even Distribution: 
Art forms like "Music", "Jewelry or other Metalworking", and "Literature" have an almost equal representation, suggesting a balanced distribution among these disciplines.

### Fewer in Film & Fashion:
 It's worth noting that domains like "Film/Television" and "Fashion or Textiles" have lower representation in the dataset. This could either indicate fewer organizations in these sectors or a lesser participation rate in the survey from such organizations.

### Specialized Niches: 
Categories like "Furniture or Woodworking" and "Graphic Design" show that there are also organizations with very specialized niches in the arts sector.

Overall, the distribution provides a comprehensive view of the varied artistic disciplines that organizations associated with "Creative Worker" and "Arts Administrator" roles focus on. This diversity is a testament to the richness and variety of the creative sector.

## Creative Workers, Arts Administrators: Provide figures and commentary regarding layoffs and the economic situation for the organization in terms of a) operating budget, and b) layoffs relative to the number of staff members.

Step 1: Filter data separately for "Creative Workers" and "Arts Administrators."

Step 2: For both categories:
a. Provide figures and visualizations regarding the operating budget.
b. Provide figures and visualizations regarding layoffs.
c. Provide figures and visualizations showing layoffs relative to the number of staff members.

Let's begin with Step 1 and filter the data separately for each group.

```python
# Filter data separately for "Creative Workers" and "Arts Administrators"
creative_workers_data = df[df["Describe Your Primary Role in Creative Sector"] == 
                                   "Creative Worker (In-house graphic designers, copywriters, art directors employed at a business working in the creative industries)"]

arts_administrators_data = df[df["Describe Your Primary Role in Creative Sector"] == 
                                      "Arts Administrator (Theater managers, development staff, executive directors, program coordinators at an arts-based business usually a non-profit, e.g., at a museum, symphony, theater, etc.)"]

# Verify the number of entries for each group
len(creative_workers_data), len(arts_administrators_data)
```

We have 50 entries each for both "Creative Workers" and "Arts Administrators."

Starting with "Creative Workers": I'll summarize and visualize the distribution of their responses for the "Annual Operating Budget of Your Organization" column.

```python
import matplotlib.pyplot as plt

# Summarize operating budget for Creative Workers
creative_workers_budget_summary = creative_workers_data["Annual Operating Budget of Your Organization"].value_counts().sort_index()

# Visualize the distribution
plt.figure(figsize=(12, 6))
creative_workers_budget_summary.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Operating Budget Distribution for Creative Workers")
plt.xlabel("Operating Budget Ranges")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Here are the figures and visualization for the operating budget distribution among "Creative Workers":

From the bar chart, we can observe the following:

The majority of "Creative Workers" are associated with organizations that operate on a budget of "All Volunteer," suggesting that many of these organizations might not have significant financial resources or traditional operational costs.
There's a fairly even distribution among the other budget ranges, with the exception of the "$1,000,000-$2,000,000" range, which has a notably lower count.

Now we provide figures and visualizations regarding the operating budget for "Arts Administrators". 

```python
# Summarize operating budget for Arts Administrators
arts_administrators_budget_summary = arts_administrators_data["Annual Operating Budget of Your Organization"].value_counts().sort_index()

# Visualize the distribution
plt.figure(figsize=(12, 6))
arts_administrators_budget_summary.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Operating Budget Distribution for Arts Administrators")
plt.xlabel("Operating Budget Ranges")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Here are the figures and visualization for the operating budget distribution among "Arts Administrators":

From the bar chart, we can observe:

A significant number of "Arts Administrators" are associated with organizations that have budgets in the "$250,001-$500,000" range, suggesting that a substantial portion of these organizations operate within this budgetary bracket.
The number of "Arts Administrators" in the "All Volunteer" budget range is notably lower compared to "Creative Workers." This could indicate that arts administration roles are more commonly found in organizations with traditional operational budgets.
Other budget ranges, such as "$0-$50,000" and "$500,001-$1,000,000", also have a considerable count, signifying the diverse financial scale of these organizations.

### Commentary on the Economic Situation Based on Operating Budget:

For both "Creative Workers" and "Arts Administrators," the spread across different budget categories suggests a wide range of organization sizes and financial capacities in the creative sector.
The presence of many entities in the lower budget ranges (including "All Volunteer") might indicate economic challenges, especially in sustaining operations without significant financial resources.

Now lets, start by examining and visualizing the layoffs situation for "Creative Workers" and then "Arts Administrators". Starting with "Creative Workers":

```python
# Summarize layoffs for Creative Workers
creative_workers_layoffs_summary = creative_workers_data["Had Employee Layoffs/Furloughs in 2020"].value_counts().sort_index()

# Visualize the layoffs distribution
plt.figure(figsize=(8, 6))
creative_workers_layoffs_summary.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Layoffs Distribution for Creative Workers in 2020")
plt.xlabel("Layoffs Status")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Here are the figures and visualization for the layoffs distribution in 2020 among "Creative Workers":

From the bar chart, we can deduce:

A majority of "Creative Workers" reported that their organizations did not have layoffs or furloughs in 2020.
However, a significant portion still experienced layoffs, indicating the economic challenges faced by some entities in this category.
Considering the economic situation:

The majority of "Creative Workers" being associated with organizations that did not resort to layoffs could imply a level of resilience or alternative strategies employed to navigate the challenges of 2020.
However, the presence of layoffs in a considerable number of organizations underscores the economic strain faced by the creative sector, especially when paired with the earlier observation of many entities operating with minimal or volunteer-based budgets.

```python
# Summarize layoffs for Arts Administrators
arts_administrators_layoffs_summary = arts_administrators_data["Had Employee Layoffs/Furloughs in 2020"].value_counts().sort_index()

# Visualize the layoffs distribution
plt.figure(figsize=(8, 6))
arts_administrators_layoffs_summary.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Layoffs Distribution for Arts Administrators in 2020")
plt.xlabel("Layoffs Status")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Here are the figures and visualization for the layoffs distribution in 2020 among "Arts Administrators":

From the bar chart, we observe:

A larger proportion of "Arts Administrators" reported that their organizations had layoffs or furloughs in 2020 compared to those that did not.
This is in contrast to "Creative Workers," where a majority did not experience layoffs.
Considering the economic situation:

The fact that more "Arts Administrators" reported layoffs compared to "Creative Workers" might suggest that organizations with traditional arts administrative roles faced greater economic challenges in 2020.
Given that many "Arts Administrators" are associated with organizations that have more substantial operating budgets, the financial strain of maintaining operations might have been more pronounced, leading to layoffs.

Now we provide figures showing layoffs relative to the number of staff members. This will help us understand the extent of the impact on staff. Starting with "Creative Workers":

```python
# Filter Creative Workers who had layoffs
creative_workers_with_layoffs = creative_workers_data[creative_workers_data["Had Employee Layoffs/Furloughs in 2020"] == "Yes"]

# Summarize staff count for Creative Workers who had layoffs
creative_workers_staff_count_summary = creative_workers_with_layoffs["Current Staff Count in Your Organization"].value_counts().sort_index()

# Visualize the staff count distribution for those who had layoffs
plt.figure(figsize=(10, 6))
creative_workers_staff_count_summary.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Staff Count Distribution for Creative Workers (Who Had Layoffs) in 2020")
plt.xlabel("Staff Count Ranges")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Here are the figures and visualization for the staff count distribution among "Creative Workers" who had layoffs/furloughs in 2020:

From the bar chart:

A significant number of "Creative Workers" associated with "All Volunteer" organizations reported layoffs. This might seem counterintuitive given the volunteer nature, but it could indicate a reduction in volunteer participation or activities in 2020.
The other prominent categories facing layoffs were organizations with staff counts in the "1-5" and "6-10" ranges. This suggests that smaller organizations might have faced challenges in retaining even their limited staff.
Given the economic situation:

Layoffs in smaller organizations and volunteer-driven entities might indicate the vulnerability of such organizations to economic downturns, as they might lack the financial buffer or resources to navigate challenges.

```python
# Filter Arts Administrators who had layoffs
arts_administrators_with_layoffs = arts_administrators_data[arts_administrators_data["Had Employee Layoffs/Furloughs in 2020"] == "Yes"]

# Summarize staff count for Arts Administrators who had layoffs
arts_administrators_staff_count_summary = arts_administrators_with_layoffs["Current Staff Count in Your Organization"].value_counts().sort_index()

# Visualize the staff count distribution for those who had layoffs
plt.figure(figsize=(10, 6))
arts_administrators_staff_count_summary.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Staff Count Distribution for Arts Administrators (Who Had Layoffs) in 2020")
plt.xlabel("Staff Count Ranges")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Here are the figures and visualization for the staff count distribution among "Arts Administrators" who had layoffs/furloughs in 2020:

From the bar chart:

A significant number of "Arts Administrators" associated with organizations that have staff counts in the "1-5" and "6-10" ranges reported layoffs. This implies that smaller arts administration entities faced considerable challenges in 2020.
Organizations with a larger staff count, such as "21-50" and "More than 50," also experienced layoffs, suggesting that even more established entities were not immune to the economic challenges.
Given the economic situation:

The fact that both smaller and larger arts administrative organizations faced layoffs underscores the widespread economic strain in 2020. Larger organizations, despite their presumably greater resources, had to resort to layoffs, possibly due to the financial burden of maintaining a bigger staff and operations.
With these insights, we've addressed the question regarding layoffs and the economic situation in terms of both the operating budget and layoffs relative to staff numbers. 

```python
# Filter data for "Creative Workers" organizations
creative_workers_layoffs_data = creative_workers_data.groupby(["Current Staff Count in Your Organization", "Had Employee Layoffs/Furloughs in 2020"]).size().unstack().fillna(0)

# Visualize the layoffs distribution relative to staff count
ax = creative_workers_layoffs_data.plot(kind='bar', stacked=True, figsize=(12, 6), color=['lightcoral', 'skyblue'], edgecolor='black')
plt.title("Layoffs Distribution Relative to Staff Count for Creative Workers in 2020")
plt.xlabel("Staff Count Ranges")
plt.ylabel("Number of Organizations")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Had Layoffs/Furloughs in 2020")
plt.tight_layout()
plt.show()
```

Here's the visualization depicting the layoffs distribution relative to staff count for "Creative Workers" organizations in 2020:

### Observations:

All Volunteer: The majority did not face layoffs, but a significant portion did. This could indicate reductions in volunteer activities or shifts in participation.
1-5 staff members: A near-even split between organizations that had and did not have layoffs.
6-10 staff members: More organizations experienced layoffs than those that didn't.
11-20 staff members: A majority did not face layoffs.
21-50 staff members: Layoffs and non-layoffs are evenly distributed.
More than 50 staff members: More organizations did not face layoffs compared to those that did.
From the chart, it's evident that the impact of layoffs in 2020 varied depending on the size of the organization. While some staff count categories showed resilience, others faced significant challenges.

```python
arts_administrators_layoffs_data = arts_administrators_data.groupby(["Current Staff Count in Your Organization", "Had Employee Layoffs/Furloughs in 2020"]).size().unstack().fillna(0)

# Visualize the layoffs distribution relative to staff count
ax = arts_administrators_layoffs_data.plot(kind='bar', stacked=True, figsize=(12, 6), color=['lightcoral', 'skyblue'], edgecolor='black')
plt.title("Layoffs Distribution Relative to Staff Count for Arts Administrators in 2020")
plt.xlabel("Staff Count Ranges")
plt.ylabel("Number of Entities")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Had Layoffs/Furloughs in 2020")
plt.tight_layout()
plt.show()
```
